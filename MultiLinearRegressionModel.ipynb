```python
import pandas as pd
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn import metrics
import numpy as np

```

```python
df = pd.read_csv("DataBase_WeatherRentalAmount_UTC.csv", index_col='FormattedDate')
```

```python
df.head()
```

```python
df.describe()
```

```python
x_train, x_test, y_train, y_test = train_test_split(df[['Temperature','PrecipitationHeight','TotalCloudCover','MeanWindSpeed','Month','Weekday','Hour']], df[['Rentals']], test_size=0.3)
```

```python
x_test
```

```python
reg = linear_model.LinearRegression()
reg.fit(x_train,y_train)
#reg.coef_
#reg.intercept_
```

```python
y_pred = reg.predict(x_test)
print(np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
Predictions = x_test
Predictions['Actual']=y_test
Predictions['Prediction']=y_pred
Predictions.head(20)
```
